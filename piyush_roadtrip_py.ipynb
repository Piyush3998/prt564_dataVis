```python
import pandas as pd

file_path = '/content/drive/MyDrive/dataAnalyticsAndVisualisation/roatrip.csv'
df = pd.read_csv(file_path)
print(df.head())
```

```python
# prompt: Plot a scatterplot to visualise the distribution of all data points of the spreadsheet

import matplotlib.pyplot as plt

# Assuming you have a DataFrame named 'df' with your data
# Select the columns you want to plot (e.g., 'x_column' and 'y_column')
x_column = 'Kilometers'  # Replace with your x-axis column name
y_column = 'Dollars'  # Replace with your y-axis column name

plt.figure(figsize=(8, 6))
plt.scatter(df[x_column], df[y_column])
plt.xlabel(x_column)
plt.ylabel(y_column)
plt.title('Scatter Plot of Distance vs. Price')
plt.grid(True)
plt.show()
```

```python
# prompt:  Build a simple linear regression model based on the dataframe's data. Write down the linear regression equation. Also, visualise the regression line over the data points

import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np

# Assuming you have a DataFrame named 'df' with your data
# Select the columns you want to use for regression
x_column = 'Kilometers'  # Replace with your x-axis column name
y_column = 'Dollars'  # Replace with your y-axis column name

X = df[[x_column]]  # Independent variable (as a DataFrame)
y = df[y_column]  # Dependent variable

# Create a linear regression model
model = LinearRegression()

# Fit the model to your data
model.fit(X, y)

# Get the coefficients and intercept
slope = model.coef_[0]
intercept = model.intercept_

# Print the Linear regression equation, Intercept and Slope
print(f"Linear Regression Equation: y = {slope:.2f}x + {intercept:.2f}")
print("Intercept: ",model.intercept_)
print("Slope: ",model.coef_)

# Predict y values using the model
y_pred = model.predict(X)

# Visualize the regression line
plt.figure(figsize=(8, 6))
plt.scatter(df[x_column], df[y_column])
plt.plot(df[x_column], y_pred, color='red', label='Regression Line')
plt.xlabel(x_column)
plt.ylabel(y_column)
plt.title('Linear Regression: Distance vs. Price')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# prompt:  Build another simple linear regression model based on the dataframe's data, but this time set the intercept to 0 (since not driving any distance at all should incur zero cost in petrol). Write down the linear regression equation and visualise the 2nd regression line over the data points

import matplotlib.pyplot as plt
# Create a linear regression model with intercept set to 0
model_no_intercept = LinearRegression(fit_intercept=False)

# Fit the model to your data
model_no_intercept.fit(X, y)

# Get the coefficient (slope)
slope_no_intercept = model_no_intercept.coef_[0]

# Print the Linear regression equation with no intercept
print(f"Linear Regression Equation (no intercept): y = {slope_no_intercept:.2f}x")
print("Intercept: ", model_no_intercept.intercept_)
print("Slope: ", model_no_intercept.coef_)

# Predict y values using the model with no intercept
y_pred_no_intercept = model_no_intercept.predict(X)

# Visualize the regression line (with no intercept)
plt.figure(figsize=(8, 6))
plt.scatter(df[x_column], df[y_column])
plt.plot(df[x_column], y_pred, color='red', label='Regression Line (with intercept)')
plt.plot(df[x_column], y_pred_no_intercept, color='green', label='Regression Line (no intercept)')
plt.xlabel(x_column)
plt.ylabel(y_column)
plt.title('Linear Regression: Distance vs. Price')
plt.grid(True)
plt.legend()
plt.show()
```
